```python
import gradio as gr
import tensorflow as tf
import numpy as np
from PIL import Image
import uuid  # For generating unique filenames


try:
  model = tf.keras.models.load_model(r"C:\Users\MRITYUNJAI RATHORE\Downloads\Data Set for ds project\galaxy10_model.h5")
except OSError as e:
  print("Error loading model:", e)
  exit() 

class_names = {
    0: 'Disturbed Galaxies',
    1: 'Merging Galaxies',
    2: 'Round Smooth Galaxies',
    3: 'In-between Round Smooth Galaxies',
    4: 'Cigar Shaped Smooth Galaxies',
    5: 'Barred Spiral Galaxies',
    6: 'Unbarred Tight Spiral Galaxies',
    7: 'Unbarred Loose Spiral Galaxies',
    8: 'Edge-on Galaxies without Bulge',
    9: 'Edge-on Galaxies with Bulge'
}


def preprocess_image(image_path, target_size=(256, 256)):

  # Load the image using PIL
  img = Image.open(image_path)

  # Resize the image to the target size
  img = img.resize(target_size)

  # Convert the image to a numpy array
  img_array = np.array(img)

  # Normalize the pixel values
  img_array = img_array.astype('float32') / 255.0

  # Expand the dimensions to match the input shape expected by the model
  img_array = np.expand_dims(img_array, axis=0)

  # Generate a unique filename for the preprocessed image
  filename = f"preprocessed_{uuid.uuid4()}.jpg"  # Adjust file format as needed

  # Save the preprocessed image to a temporary location
  Image.fromarray(img_array.astype('uint8')).save(filename)

  return img_array, img, filename


def predict_class(model, image_path):

  # Preprocess the image
  preprocessed_image, _, _ = preprocess_image(image_path)

  # Make prediction
  prediction = model.predict(preprocessed_image)

  # Get the predicted class index
  predicted_class = np.argmax(prediction)

  return predicted_class, image_path


def launch_gradio_interface():

  
  interface = gr.Interface(
      fn=lambda image: (
          predict_class(model, image)[0],
          {class_name: float(probability) for class_name, probability in zip(class_names.values(), model.predict(preprocess_image(image)[0])[0])}
      ),
      inputs=gr.Image(label="Select galaxy image"),
      outputs=[gr.Label(num_top_classes=3), gr.Text(label="Original Image")], 
      title="Galaxy Image Classification"
  )

  interface.launch()



launch_gradio_interface()
```
